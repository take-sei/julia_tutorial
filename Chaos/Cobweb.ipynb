```julia
using Plots
gr()
```

```julia
include("./Utility/Cobweb.jl")
using Cobweb
```

```julia
# Logistic funtion
f(x) = 2x * (1-x)
# linear function
g(x) = 2x

# f:domein -> range
domein = [0.01i for i in 0:100]
range = [f(i) for i in domein]

# Diagonal(x) = x
diagonal = domein |> copy;
```

```julia
x,y = render(0.01, g, 7)
plot(domein, [[g(i) for i in domein], diagonal])
plot!(x, y, label="")
```

```julia
x, y = render(0.01, f)
plot(domein,[range, diagonal])
plot!(x, y, label="")
```
