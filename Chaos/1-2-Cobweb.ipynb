```julia
# Case 1
# Linear function: f(x) = 2x
# Initial Value: x_0 = 0.01
# Max Trial: N = 10

f(x) = 2x
x_0 = 0.01
N = 10

Domein = [0.01i for i=0:100]
cobweb_plot(f, Domein, x_0, N)
```

```julia
# Case 2
# Logistic function: g(x) = 2x * (1-x)
# Initial Value: x_0 = 0.01
# Max Trial: N = 10

g(x) = 2x * (1-x)
x_0 = 0.01
N = 10

Domein = [0.01i for i=0:100]
cobweb_plot(g, Domein, [0.01, 0.8], N)
```

```julia
# Case 3
# Cubic function: h(x) = (3x-x^3) / 2
# Initial Value: x_0 = 0.01
# Max Trial: N = 10

h(x) = (3x-x^3) / 2
x_0 = 0.01
N = 10

Domein = [0.01i for i=-200:200]
cobweb_plot(h, Domein, [1.6, 1.8, 0.01])
```
