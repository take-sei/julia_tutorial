# Periodic Point
ロジスティック写像のパラメータをいじることで, 周期点を確認する.

- $f(x) = \alpha x(1-x)$
- パラメータ: $\alpha$

```julia
using Plots 
gr()

using TypedTables

include("./Utility/Cobweb.jl")
using Cobweb
```

```julia
logistic(a, x) = a*x * (1-x)

f(x) = logistic(3.3, x)
x_0 = [0.2, 0.5, 0.95]
N = 14
```

```julia
orbits = []
label = []

for x in x_0
  passway = zeros(N+1)
  passway[1] = x
  
  for i=1:N
    passway[i+1] = f(passway[i])
  end
  
  push!(orbits, passway)
  push!(label, "a=$x")
end
```

```julia
plot(collect(0:N), orbits, label=label)
```

```julia
Domein = [i for i=0:0.01:1]
cobweb_plot(f, Domein, x_0, N)
plot!()
```
